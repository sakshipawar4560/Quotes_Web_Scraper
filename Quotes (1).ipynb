```python
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url="https://quotes.toscrape.com/"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup
```

```python
table=soup.find("body")
table=soup.find("h1")
table
```

```python
r=soup.find(class_='row')
r=soup.find_all(class_="text")
r1=soup.find_all(class_="author")
print(f"{r} - {r1}")
# print(r1)
```

```python
q=soup.find_all("span" ,class_="text")
q
```

```python
a=soup.find_all("small" , class_="author")
a
```

```python
qs=soup.find_all("div" , class_="quote")
qs
```

```python
quotes_info=[]
Quotes=[]
Authors=[]
for i in qs:
  # cols=i.text
  cols=i.find("span" , class_="text")
  rows=i.find("small", class_="author")
  # rows=
  # print(cols)
  # print(rows)
  # print(f"{cols} - {rows}")
  Quotes.append(cols)
  Authors.append(rows)

```

```python
print(Quotes)
print(Authors)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Quote": Quotes,
    "Author": Authors
})


print(df.head())
```

```python
df.to_csv("Quotes scrape.csv")
```
